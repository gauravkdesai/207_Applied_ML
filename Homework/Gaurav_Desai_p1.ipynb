The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:

```python
def P6():
    
### STUDENT START ###
    blurred_mini_train_data = blurImageData(mini_train_data)
    blurred_dev_data = blurImageData(dev_data)
    
    k=1
    
    print("Using k=",k)
    print("Without blurring anything")
    runKnnWithblurredData(k=k,my_train_data=mini_train_data, my_train_labels=mini_train_labels,
                                 my_dev_data=dev_data,my_dev_labels=dev_labels )
    
    print("After blurring training data")
    runKnnWithblurredData(k=k,my_train_data=blurred_mini_train_data, my_train_labels=mini_train_labels,
                                 my_dev_data=dev_data,my_dev_labels=dev_labels )
    
    print("After blurring dev data")
    runKnnWithblurredData(k=k,my_train_data=mini_train_data, my_train_labels=mini_train_labels,
                             my_dev_data=blurred_dev_data,my_dev_labels=dev_labels )
    
    print("After blurring both training data and dev data")
    runKnnWithblurredData(k=k,my_train_data=blurred_mini_train_data, my_train_labels=mini_train_labels,
                                 my_dev_data=blurred_dev_data,my_dev_labels=dev_labels )    
    
def runKnnWithblurredData(k,my_train_data,my_train_labels,my_dev_data,my_dev_labels):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(my_train_data,my_train_labels)
    my_dev_predict=knn.predict(my_dev_data)
    cr=classification_report(my_dev_labels,my_dev_predict)
    print(cr)
    
 
def blurImageData(originalImageData):
    blurredImageData=np.empty((len(originalImageData),28*28))
    for i in range(len(originalImageData)):
        blurredImageData[i]=simpleBlurImage(originalImageData[i].reshape(28,28)).ravel()
    return blurredImageData

def simpleBlurImage(originalImage):
    blurredImage=np.copy(originalImage)
    wt5=.50 # Half weight for original cell value
    wt1=wt2=wt3=wt4=wt6=wt7=wt8=wt9=((1-wt5)/8)    # Rest of the weight distributed to adjacent 8 cells
    
    maxRow=originalImage.shape[0]
    maxColumn=originalImage.shape[0]
    for i in range(maxRow):
        for j in range(maxColumn):
            c1=wt1 * originalImage[i-1][j-1] if i>0 and j>0 else 0
            c2=wt2 * originalImage[i-1][j] if i>0 else 0
            c3=wt3 * originalImage[i-1][j+1] if i>0 and j<maxColumn-1 else 0
            
            c4=wt4 * originalImage[i][j-1] if j>0 else 0
            c5=wt5 * originalImage[i][j] 
            c6=wt6 * originalImage[i][j+1] if j<maxColumn-1 else 0
            
            c7=wt7 * originalImage[i+1][j-1] if i<maxRow-1 and j>0 else 0
            c8=wt8 * originalImage[i+1][j] if i<maxRow-1 else 0
            c9=wt9 * originalImage[i+1][j+1] if i<maxRow-1 and j<maxColumn-1 else 0
            
            blurredImage[i][j] = c1+c2+c3+c4+c5+c6+c7+c8+c9
            
    return blurredImage            
        
    

### STUDENT END ###

P6()
```

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    
### STUDENT START ###
        
    clf = BernoulliNB(alpha=0.0001)
    clf.fit(mini_train_data,mini_train_labels)
    dev_predicts = clf.predict(dev_data)
    dev_proba=clf.predict_proba(dev_data)

    for i in range(len(dev_predicts)):
        #print("i=",i)
        #print("dev_data=",dev_data[i])
        #print("dev_predicts=",dev_predicts[i])
        #print("dev_proba=",dev_proba[i])

        maxProbIndex = np.argmax(dev_proba[i])
        maxProb = dev_proba[i][maxProbIndex]
        
        #print("maxProbIndex",maxProbIndex,"maxProb",maxProb)
        
        for b in range(len(buckets)):
            if maxProb <= buckets[b] :
                total[b] = total[b] + 1
                if dev_predicts[i] == dev_labels[i] :
                    correct[b] = correct[b] + 1
                break
                
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i] / total[i]
    print('p(pred) <= %.13f    total = %3d    correct = %3d    accuracy = %.3f' %(buckets[i], total[i], correct[i], accuracy))
```

ANSWER:

The Naive Bayes model appears to be weakly caliberated as acurancy increases wrt increase in posterior probability
